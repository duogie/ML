```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('penguins.csv')
df = df[(df['flipper_length_mm'] != 5000.0)]
df = df.dropna()
```

```python
df.info()
```

```python
import seaborn as sns
sns.scatterplot(data=df, x='culmen_length_mm', y='body_mass_g')
```

```python
from sklearn.model_selection import train_test_split
train_df, _ = train_test_split(df, test_size=0.3, random_state=42)
train_df = train_df[["body_mass_g", "culmen_length_mm", "flipper_length_mm"]]
train_df.head()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
model.fit(train_df)
```

```python
model.cluster_centers_
```
